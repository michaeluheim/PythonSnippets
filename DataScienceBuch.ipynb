```python
users = [
  { "id": 0, "name": "Hero" },
  { "id": 1, "name": "Alice Wonderland" },
  { "id": 2, "name": "Bob Builder" },
  { "id": 3, "name": "Charlie Checkpoint" },
  { "id": 4, "name": "Diana Dash" },
  { "id": 5, "name": "Evan Example" },
  { "id": 6, "name": "Fiona Focus" },
  { "id": 7, "name": "George Generic" },
  { "id": 8, "name": "Hannah Help" },
  { "id": 9, "name": "Ian Input" },
  { "id": 10, "name": "Julia JSON" }
];

friendship_pairs = [
    (0, 1), (0, 2), # Originale
    (0, 4), (0, 5), (0, 8), (0, 9), (0, 10), 
    (1, 2), (1, 7), (1, 8), (1, 9), 
    (2, 3), 
    (3, 6), 
    (4, 5), (4, 7), (4, 10), 
    (5, 8), (5, 9), 
    (6, 7), (6, 10), 
    (7, 8), 
    (8, 9)
]
```

```python
friendships = { user["id"]: [] for user in users }
for i, j in friendship_pairs:
    friendships[i].append(j)
    friendships[j].append(i)

friendships
```

```python
def number_of_friends(user):
    #wie viele Freunde hat User?
    user_id = user["id"]
    friends_ids = friendships[user_id]
    return len(friends_ids)

total_connections = sum(number_of_friends(user) for user in users)
num_users = len(users)

avg_connections = num_users/total_connections
avg_connections

#Liste aller User mit Anzahl Freunde
number_of_friends_by_id = [(user["id"],number_of_friends(user)) for user in users]

#lamda geht auf das 1te element, also hier Anzahl der Freunde
#id_and_friends = (1, 5)
#id_and_friends[1] = 5
number_of_friends_by_id.sort(key = lambda id_and_friends: id_and_friends[1], reverse=True)
number_of_friends_by_id

```
